```python
import csv
import requests
import argparse

# 读取CSV文件的两列
def read_csv(csv_file):
    with open(csv_file, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        data = [(row[0], row[1]) for row in reader]  # 读取问题列和标准答案列
    return data

# 发送请求到模型并获取回答
def get_model_response(prompt, model_name):
    messages = [{"role": "user", "content": prompt}]
    prompt_headers = {"Authorization": "Bearer your_api_key_here"}
    stream_mode = False
    data = {
        "model": model_name,  # 使用参数化的模型名称
        "stream": stream_mode,
        "messages": messages,
    }

    response = requests.post(
        "http://127.0.0.1:8000/v1/chat/completions",
        headers=prompt_headers,
        json=data,
        stream=stream_mode,
    )
    return response.json()

# 将结果写入CSV文件
def write_results_to_csv(output_file, results, model_name):
    # 检查文件是否存在，如果不存在则写入表头
    try:
        with open(output_file, mode='r', newline='', encoding='utf-8') as file:
            reader = csv.reader(file)
            headers = next(reader)  # 读取现有表头
    except FileNotFoundError:
        headers = ["问题", "标准答案"]

    # 如果模型相关的列不存在，则追加到表头
    model_columns = [f"{model_name}", f"{model_name}_message", f"{model_name}_completion_tokens"]
    for column in model_columns:
        if column not in headers:
            headers.extend(model_columns)
            break

    # 写入数据
    with open(output_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(headers)  # 写入更新后的表头
        for result in results:
            row = [result['问题'], result['标准答案']]
            if model_name in result:
                row.extend([result[model_name], result[f"{model_name}_message"], result[f"{model_name}_completion_tokens"]])
            else:
                row.extend(["", "", ""])  # 如果模型结果不存在，填充空值
            writer.writerow(row)

# 主函数
def main(input_csv, output_csv, model_names):
    data = read_csv(input_csv)  # 读取输入CSV文件

    results = []
    for question, answer in data:
        result = {
            '问题': question,
            '标准答案': answer
        }
        results.append(result)

    # 依次测试每个模型
    for model_name in model_names:
        for result in results:
            prompt = result['问题']
            response = get_model_response(prompt, model_name)
            message = response['choices'][0]['message']['content']
            completion_tokens = response['usage']['completion_tokens']

            # 将模型结果添加到当前行
            result[model_name] = model_name
            result[f"{model_name}_message"] = message
            result[f"{model_name}_completion_tokens"] = completion_tokens

        # 将当前模型的结果写入CSV文件
        write_results_to_csv(output_csv, results, model_name)

    print(f"Results have been written to {output_csv}")

if __name__ == "__main__":
    # 设置命令行参数解析
    parser = argparse.ArgumentParser(description="Process CSV and get model responses.")
    parser.add_argument('--input_csv', type=str, required=True, help="Path to the input CSV file.")
    parser.add_argument('--output_csv', type=str, required=True, help="Path to the output CSV file.")
    parser.add_argument('--model_names', type=str, nargs='+', required=True, help="Names of the models to test.")

    # 解析命令行参数
    args = parser.parse_args()

    # 调用主函数
    main(args.input_csv, args.output_csv, args.model_names)
```
